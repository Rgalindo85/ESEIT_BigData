```python
raw_data = get_input_data(
    bucket=bucket,
    initial_directory=directory,
    filename= 'datos_abiertos_cons_diciembre_2019.csv',
)
print(raw_data.info())
```

## Renombras Columnas

* ) leemos la base para validar los campos

```python
raw_data.rename(
    columns = {
      
        'NUMERO_INCIDENTE'                  : 'NUMERO_INCIDENTE',
        'FECHA_DESPACHO_518'                : 'FECHA_INICIO_DESPLAZAMIENTO_MOVIL',
        'CODIGO DE LOCALIDAD'               : 'CODIGO_LOCALIDAD',
        'LOCALIDAD'                         : 'LOCALIDAD',
        'EDAD'                              : 'EDAD',
        'UNIDAD'                            : 'UNIDAD',
        'GENERO'                            : 'GENERO',
        'RED'                               : 'RED',
        'TIPO_INCIDENTE'                    : 'TIPO_DE_INCIDENTE',
        'PRIORIDAD'                         : 'PRIORIDAD',
        'MES'                               : 'MES',
        'CLASIFICACION FINAL'               : 'CLASIFICACION_FINAL',
      
    },
    inplace=True
)
raw_data.head()
```

```python
del(raw_data['Unnamed: 36'])
```

# Valores por Campo

* ) Aplicamos Funcion Lambda

*) NUMERO_INCIDENTE

```python
col = 'NUMERO_INCIDENTE'
raw_data[col].fillna('SIN_DATO', inplace=True)
raw_data[col] = raw_data[col].apply(lambda x: x.strip())
```

* ) Fecha inicio desplazamiento Movil

```python
col = 'FECHA_INICIO_DESPLAZAMIENTO_MOVIL'
raw_data[col] = pd.to_datetime(raw_data[col], format='%Y/%m/%d %H:%M') 
raw_data[col].value_counts
```

```python
col = 'FECHA_INCIDENTE'
raw_data[col] = pd.to_datetime(raw_data[col], format='%Y/%m/%d %H:%M') 
raw_data[col].value_counts
```

* ) Edad

```python
col = 'EDAD'
raw_data[col].fillna('SIN_DATO', inplace=True)
raw_data[col] = raw_data[col].apply(lambda x: x.strip())
```

* ) Unidad

```python
col = 'UNIDAD'
raw_data[col].fillna('SIN_DATO', inplace=True)
raw_data[col] = raw_data[col].apply(lambda x: x.strip())
```

* ) Localidad

```python
raw_data['LOCALIDAD'].value_counts(dropna=False)
raw_data['LOCALIDAD'] = raw_data['LOCALIDAD'].apply(lambda x: x.strip())
raw_data['LOCALIDAD'].value_counts(dropna=False)
```

* ) Genero

```python
col = 'GENERO'
raw_data[col].fillna('SIN_DATO', inplace=True)
raw_data[col] = raw_data[col].apply(lambda x: x.strip())
```

* )  Tipo Incidente

```python
col = 'TIPO_DE_INCIDENTE'
raw_data[col].fillna('SIN_DATO', inplace=True)
raw_data[col] = raw_data[col].apply(lambda x: x.strip())
```

* ) RED

```python
col = 'RED'
raw_data[col].fillna('SIN_DATO', inplace=True)
raw_data[col] = raw_data[col].apply(lambda x: x.strip())
raw_data.head()
```

* ) Prioridad

```python
col = 'PRIORIDAD'
raw_data[col].fillna('SIN_DATO', inplace=True)
raw_data[col] = raw_data[col].apply(lambda x: x.strip())
raw_data.head()
```
