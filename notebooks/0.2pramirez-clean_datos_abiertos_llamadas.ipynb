# Check values per field

```python
llamadas_data['LOCALIDAD'].value_counts(dropna=False)
```

```python
def normalize(s):
    replacements = (
        ("á", "a"),
        ("é", "e"),
        ("í", "i"),
        ("ó", "o"),
        ("ú", "u"),
    )
    for a, b in replacements:
        s = s.replace(a, b).replace(a.upper(), b.upper())
    return s
```

```python
llamadas_data['LOCALIDAD']= llamadas_data['LOCALIDAD'].apply(
    lambda x:normalize(x)        
)
llamadas_data['LOCALIDAD'].value_counts(dropna=False, ascending=True)
```

```python
# Let's use the function strip(), this property works for string objects in python that removes whitespaces
llamadas_data['LOCALIDAD'].apply(
    lambda x: x.strip()        # here we use a lambda function, exclusively for python. Is a one-line function
).value_counts()
```

```python
llamadas_data['LOCALIDAD'] = llamadas_data['LOCALIDAD'].apply(lambda x: x.strip()) # we assign the output of the function apply to the same column
```
