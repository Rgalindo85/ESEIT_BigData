```python
list_cols_to_drop=['Unnamed: {}'.format(i) for i in range(10,37)]
list_cols_to_drop
#Eliminar columna de vacios (si axis=1 elimina columnas, si axis=0 elimina)
data.drop(list_cols_to_drop,axis=1).info()
```

```python
data_master=data.drop(list_cols_to_drop,axis=1)
data_master
```

```python
col='FECHA_INCIDENTE'
data_master[col] = pd.to_datetime(data_master[col])
data_master
```

```python
col='FECHA_INICIO_DESPLAZAMIENTO_MOVIL'
data_master[col].fillna('SIN_DATO').value_counts()
data_master[col].fillna('SIN_DATO', inplace=True)
#data_master[col] = data_master[col].apply(lambda x: x.strip())
data_master
```

```python
col='FECHA_INCIDENTE'
data_master[col].fillna('SIN_DATO').value_counts()
data_master[col].fillna('SIN_DATO', inplace=True)
#data_master[col] = data_master[col].apply(lambda x: x.strip())
data_master
```

```python
data_master
```

```python
col='MES'
data_master[col].fillna('SIN_DATO').value_counts()
data_master[col].fillna('SIN_DATO', inplace=True)
data_master[col] = data_master[col].apply(lambda x: x.strip())
data_master
```

```python
col='CLASIFICACION_FINAL'
data_master[col].fillna('SIN_DATO').value_counts()
data_master[col].fillna('SIN_DATO', inplace=True)
data_master[col] = data_master[col].apply(lambda x: x.strip())
data_master
```

```python
col='NUMERO_INCIDENTE'
data_master[col].fillna('SIN_DATO').value_counts()
```

```python
data_master[col].fillna('SIN_DATO', inplace=True)
data_master
```

```python
col='FECHA_DESPACHO_518'
data_master[col].fillna('SIN_DATO').value_counts()
data_master[col].fillna('SIN_DATO', inplace=True)
data_master[col] = data_master[col].apply(lambda x: x.strip())
data_master
```

```python
data_master=data_master[['NUMERO_INCIDENTE','FECHA_INCIDENTE','FECHA_DESPACHO_518','FECHA_INICIO_DESPLAZAMIENTO_MOVIL','CODIGO_LOCALIDAD','EDAD','UNIDAD','LOCALIDAD','GENERO','RED','TIPO_INCIDENTE','PRIORIDAD','MES','CLASIFICACION_FINAL']]
data_master
```
